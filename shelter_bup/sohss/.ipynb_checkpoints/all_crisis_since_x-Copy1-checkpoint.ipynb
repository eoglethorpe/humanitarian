```python
#pulling data

def pull_indiv(data):
    """
    return all individual crisis data, exclude ones that don't have a 200 status code
    """
    global resps
    hrefz = [h['href'] for h in data]

    def exception_handler(request, exception):
         print('Bad URL for ' + request)

    resps = []
    it = 200
    for v in range(0, len(hrefz), it):
        print('Pulling individual for hrefs to ' + str(v))
        rs = (grequests.get(ref) for ref in hrefz[v : v+it])
        resps += grequests.map(rs, exception_handler = exception_handler)
    
    return [json.loads(r.content) for r in resps if r.status_code == 200]

def merge(ind, data):
    

def fetch_api(ind, data):
    """
    pull down all API info for general crisis and then crisis specific, make into a dict
    """
    global data
    data = []
    for i in range(4):
        with urllib.request.urlopen("https://api.reliefweb.int/v1/disasters?appname=vocabulary"
                                    "&preset=external&limit=1000&offset={}".format(i*1000)) as url:
            data += json.loads(url.read().decode())['data']

    return merge(pull_indiv(data), data)
```

```python
def get_date(val):
    name = val['fields']['name'].replace(' ', '')[-7:]
    month = None
    year = None
    
    MIN_YEAR = 2005
    #if we don't have regular formatting, take just year
    if not name[0:3].isalpha():
        if not name[-4:].isnumeric():
            print("bad year: " + name)
        else:
            year = name[-4:]
    else:
        month = name[:3]
        year = name[-4:]
    
    if year:
        if int(year) >= MIN_YEAR:
            return month, year

def trim_nm(nm):
    s_val = None
    
    if len(nm.split('-')) != 1:
        s_val = '-'
    elif len(nm.split('–')) != 1:
        s_val = '–' 
    
    if s_val:
        return nm.split(s_val)[0]
    else:
        return nm
    
def get_schwifty(v):
    """
    get specific info for disaster
    """
    try:
        with urllib.request.urlopen(v['href']) as url:
            info = json.loads(url.read().decode())
            return info
    except:
        print('Bad specific info for: ' + str(v))
#         print(len(info['data']['country']))
        
#         data += json.loads(url.read().decode())['data']



def master_pull():
    """take crises only after certain year, add month_crisis: mmm, and year_crisis: yyyy to each crisis's json

        names are either in format of:
            MMM YYYY
            OR
            YYYY-YYYY

        if not in first format, check to see if end year > 2005
    """
    
    #only pull data if not already present 
    if 'has_run' not in globals().keys():
        
    if 'data' not in globals().keys():
        data = fetch_api()
    else:
        data = globals()['data']
        
    filt_data = []
    c = 0
    for v in data:
        if get_date(v):
            month, year = get_date(v)
            v['fields']['month_crisis'] = month
            v['fields']['year_crisis'] = year
            filt_data.append(v)
        
        #external calls
        v['fields']['name'] = trim_nm(v['fields']['name'])
#         v['specific'] = get_schwifty(v)
    
    
    return filt_data
```

```python
def write():
    d = master_pull()
    with open('crises.csv', 'w') as csv_file:
        out = csv.writer(csv_file, delimiter=',')
        out.writerow(('crisis_nm', 'date_month', 'date_year', 'glide', 'url'))
        for v in d:
            out.writerow([v['fields']['name'], v['fields']['month_crisis'], v['fields']['year_crisis'], v['fields']['glide'], v['href']])
            
write()
```
