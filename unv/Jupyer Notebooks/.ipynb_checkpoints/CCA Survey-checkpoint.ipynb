```python
df = df.drop(1)
df = df.drop(2)
df = df.drop(0)
```

```python
#add work type and in/out columns

df = df.drop(df.columns[[1,2,3,4,5,6,7,8]], axis=1)
df.insert(1, 'work_type', None)

df.insert(2, 'location', None)

```

```python
#find out location

def not_blank(inp):
    if len(''.join([str(i) for i in inp])) >= 1:
        return True
    else:
        return False

def make_loc(row):
    if not_blank(row[[i for i in range(3,24,2)]]):
        return 'Central'
    elif not_blank(row[[i for i in range(4,25,2)]]): 
        return 'Outside'
    else:
        return 'Blank'

df['location'] = df.apply (lambda row: make_loc(row),axis=1)

g = df.groupby('location')

g.size()
```

```python
#find out type working for

# Government of Nepal - Working at central level	Government of Nepal - Working at sub-national, district/VDC level
# International NGO - Working at central level	International NGO - Working at sub-national, district/VDC level	
# National / Local NGO - Working at central level	National / Local NGO - Working at sub-national, district/VDC level
# Donors - Working at central level	Donors - Working at sub-national, district/VDC level	
# Development Partners - Working at central level	Development Partners - Working at sub-national, district/VDC level	
# UN - Working at central level	UN - Working at sub-national, district/VDC level	
# Private Sector - Working at central level	Private Sector - Working at sub-national, district/VDC level	
# Academic or Research Institution - Working at central level	Academic or Research Institution - Working at 
# Beneficiariesâ€™ Federations /Associations - Working at central level	Beneficiariesâ€™ Federations 
# Media - Working at central level	Media - Working at sub-national, district/VDC level	
# Others - Working at central level	Others - Working at sub-national, district/VDC level

def make_org(row):
    if row[2] or row[3] != '':
        return 'GoN'
    elif row[4] or row[5] != '':
        return 'INGO'
    elif row[6] or row[7] != '':
        return 'Local NGO'
    elif row[8] or row[9] != '':
        return 'Donor'
    elif row[10] or row[11] != '':
        return 'Dev Partner'
    elif row[12] or row[13] != '':
        return 'UN'
    elif row[14] or row[15] != '':
        return 'Private'
    elif '' not in row[16:23].values:
        return 'Other'
    
df['work_type'] = df.apply (lambda row: make_org(row),axis=1)

df.groupby('work_type').groups
```

```python
# [(x,i) for x,i in enumerate(df.columns[1:30])]

d = df.iloc[0][16:23]

'' not in d.values
```

```python
#work location
wl = df[df.ix[0,df.ix[0] == 'work_location '].index.tolist()]

wl = wl.replace([1,2],1)

#add cent, ext
centext = np.tile(['central','exterior'], 11)
centext = np.hstack((centext, np.array(' ')))

def add_row(row, df):
    df.loc[-1] = row
    df.index = df.index + 1
    df = df.sort()
    return df
    
wl = add_row(centext, wl)

#group
grouped = ["GoN","GoN", "INGO","INGO", "Local NGO","Local NGO", "Donor","Donor", "Dev Partner","Dev Partner", "UN","UN", "Private","Private", "Other","Other","Other","Other","Other","Other","Other","Other","Other"]

wl = add_row(grouped, wl)
wl
```
